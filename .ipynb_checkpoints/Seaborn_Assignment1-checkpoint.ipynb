## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Q4: Add a new_column to a DataFrame that combines class and embark_town

```python
df['new_column'] = df['class'].str.cat(df['embark_town'], sep='-')
df.head()
```

# Q5: Remove new_column from the DataFrame

```python
df= df.drop('new_column',axis='columns')
df.head()
```

# Q6: Filter DataFrame for rows of survived Males only 

```python
df[(df['survived'] == 1) & (df['sex'] == 'male') ] 
```

# Q7: The total number of males who survived 

```python
males = df[(df['survived'] == 1) & (df['sex'] == 'male') ]
# males.shape to check number of rows which is identical to the below line output
print('number of males who survived : ',males['survived'].count())
```

# Import Seaborn

```python
import seaborn as sns
import numpy as np
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(x='sex', y='survived',data=df);
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(
data=df,
x='embarked',
y='survived',
kind='bar',
height=4,
alpha=.8
);
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
sns.boxplot(x='age',y='sex',hue='pclass',data=df);
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
corr = df[['sibsp','parch','age','fare','survived']].corr()
sns.heatmap(corr,annot=True);
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
#as the factorplot has been removed from sns library, i used catplot to show the relationship between sibsp and survived columns
# Please follow this link as it explains the new version of factorplot https://stackoverflow.com/questions/75207721/change-the-factorplot-of-seaborn-to-include-dots

sns.catplot(
data=df,
x='sibsp',
y='survived',
kind='bar',
height=3,
alpha=.8
);


```
